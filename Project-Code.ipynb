# HYPOTHESIS TESTING

```python
#Q1
#H0: There is no significant difference in the mean salary between male and female with the same educational level
#HA: There is a significant difference in the mean salary between male and female with the same educational level

#Q2
#H0: There is no significant difference in the mean salary between male and female with the same job description
#HA: There is a significant difference in the mean salary between male and female with the same job description

#Q3
#H0:Age and gender are independent
#HA: Age and gender are dependent
```

# EDUCATION - DOCTORAL & MASTERS (PT)

```python
education_data = portugal_data[portugal_data['Highest Education'] =='Doctoral degree']
education_data.head()
```

```python
#DOCTORAL DEGREE
contingency_table = pd.crosstab(education_data['Gender'], education_data['Salary'])

chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")

#we cannot not reject the H0
```

```python
education_data2 = portugal_data[portugal_data['Highest Education'] =='Master’s degree']
education_data2.head()
```

```python
#MASTERS DEGREE
contingency_table = pd.crosstab(education_data2['Gender'], education_data2['Salary'])

chi2, p, dof, expected = chi2_contingency(contingency_table)


print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p}")

#we cannot not reject the H0
```
